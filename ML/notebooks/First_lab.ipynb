```python
def model_one_prep(new_train):
    new_train = new_train[['id', 'full_sq','life_sq', 'floor', 'max_floor', 'num_room', 'theater_km', 'metro_km_walk', 'kremlin_km']]

    new_train['life_sq'].fillna(new_train['full_sq']*0.75, inplace = True)

    new_train['theater_km'].fillna(0, inplace = True)

    new_train['metro_km_walk'].fillna(0, inplace = True)

    new_train['kremlin_km'].fillna(0, inplace = True)

    new_train['floor'].fillna(int(new_train['floor'].mean()), inplace = True)

    new_train['max_floor'].fillna(int(new_train['max_floor'].quantile(0.75)), inplace = True)

    new_train['num_room'].fillna(int(new_train['num_room'].mean()), inplace = True)
    return new_train
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_squared_log_error, mean_absolute_error
from sklearn.linear_model import Ridge
```

```python
def train_model(data, target, features, n_splits = 5):
    msle_list = []
    mse_list = []
    mae_list = []
    
    kf = KFold(n_splits = n_splits, shuffle = True)
    for i, (train_index, val_index) in enumerate(kf.split(data)):
        train_part = data.loc[train_index,:]
        val_part = data.loc[val_index,:]
        
        y_train = target[train_index]
        y_val = target[val_index]

#         model = SGDRegressor(loss = 'huber')
        model = SGDRegressor()
#         model = Ridge(alpha = 1.0)
        model.fit(X=train_part[features], y = y_train )
        val_prediction = model.predict(val_part[features]).clip(0, 1000000000)
        
        print(val_prediction)
        
        mae = mean_absolute_error(val_prediction,  y_val)

        mse = mean_squared_error(val_prediction,  y_val)
#         msle = mean_squared_log_error(val_prediction,  val_part[target])


#         msle_list.append(msle)
        mse_list.append(mse)
        mae_list.append(mae)
        print( f'Fold {i}:  mse {mse}, mae {mae}')
        
#     print(f'MSLE average = {np.mean(msle_list)} , std ={np.std(msle_list)}  ')
    print(f'MSE average = {np.mean(mse_list)} , std ={np.std(mse_list)}  ')
    print(f'MAE average = {np.mean(mae_list)} , std ={np.std(mae_list)}  ')
    return model
```
