```python
from natasha import Doc, MorphVocab, NewsNERTagger, NewsEmbedding, Segmenter
```

```python
import pandas as pd
```

```python
df = pd.read_csv('text_series_full.csv')
```

```python
pd.options.display.max_colwidth = 500
df.head()
```

```python
emb = NewsEmbedding()
ner_tagger = NewsNERTagger(emb)
segmenter = Segmenter()
```

```python
text = df.iloc[4,1]
doc = Doc(text)
```

```python
 doc.segment(segmenter)
```

```python
doc.tag_ner(ner_tagger)
print(doc.spans[:5])
doc.ner.print()
```

```python
for i in range(1, 100, 10):
    text = df.iloc[i,1]
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_ner(ner_tagger)
    doc.ner.print()
```
